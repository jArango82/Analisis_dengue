| `dengue_label`                    | Etiqueta que indica si el paciente tiene dengue (1) o no (0).               | Numérico     |

# Manejar los valors Faltantes

## Imputar o eliminar valorez faltantes



```python
# Imputar o eliminar valorez faltantes
df['wbc_count'].fillna(df['wbc_count'].median(), inplace=True)
df['platelet_count'].fillna(df['platelet_count'].median(), inplace=True)
df['platelet_distribution_width'].fillna(df['platelet_distribution_width'].median(), inplace=True)
```

```python
# Cantidad de valores faltantes despues de la imputación:
print("Valores faltantes por columna después de la imputación:")
print(df[['wbc_count', 'platelet_count', 'platelet_distribution_width']].isnull().sum())
```
